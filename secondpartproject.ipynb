```python
import pandas as pd
import numpy as np

from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"

df_teams = pd.read_csv("teams.csv")
df_regular_season_2015 = pd.read_csv("regular_season_compact_results_2015.csv")
```

```python
np_teams = df_teams["team_id"].to_numpy()
np_results = df_regular_season_2015.to_numpy()
```

```python
##find wp wining percentage
listofoponents = []
data = []
for x in np_teams:
    # print(x)
    wincounter = 0;
    numofgames = 0;
    listofoponents = []
    for y in np_results:
        # print("win", y[2])
        if int(y[2]) == x:
            # print("++++++++++")
            listofoponents.append(int(y[4]))
            wincounter = wincounter + 1
            numofgames = numofgames + 1
        # print("lose", y[4])
        if int(y[4]) == int(x):
            # print("+++++++++++")
            listofoponents.append(int(y[2]))
            numofgames = numofgames + 1

    # print("win", wincounter)
    # print("all", numofgames)
    # print(x)

    ##odbacene ekipe koje nisu odigrale nijednu utakmicu u sezoni
    if(numofgames != 0):
        win_percentage = wincounter / numofgames
        data.append( [x, win_percentage, listofoponents])

    # print(df_teams[df_teams["team_id"] == int(x)]["team_name"])
    # print(win_percentage)
    ##print("--------")


df_win_percentage = pd.DataFrame(data, columns=["team_id", "win_percentage", "list_oponents"])

df_win_percentage





```

```python
##find op
np_win_percentage = df_win_percentage.to_numpy()

data = []
for x in np_win_percentage:
    counter = len(x[2])
    opp_percentage = 0
    for y in x[2]:
        for z in np_win_percentage:
            if int(z[0]) == int(y):
                opp_percentage = opp_percentage + float(z[1])

    opp_percentage = opp_percentage / counter
    print("team", x[0])
    print("oponents", counter)
    print("opp_percentage", opp_percentage)
    data.append(opp_percentage)

df_win_percentage["opp_percentage"] = data

df_win_percentage
```

```python
##find oop
np_win_percentage = df_win_percentage.to_numpy()
data = []
for x in np_win_percentage:
    counter = len(x[2])
    opp_opp_percentage = 0
    for y in x[2]:
        for z in np_win_percentage:
            if int(z[0]) == int(y):
                opp_opp_percentage = opp_opp_percentage + float(z[3])


    opp_opp_percentage = opp_opp_percentage / counter
    # print("team", x[0])
    # print("oponents", counter)
    # print("opp_opp_percentage", opp_opp_percentage)
    data.append(opp_opp_percentage)

df_win_percentage["opp_opp_percentage"] = data

df_win_percentage
```

```python
## calculate rpi
data = []
np_calc_rpi = df_win_percentage.to_numpy()
for x in np_calc_rpi:
    wp = float(x[1])
    op = float(x[3])
    oop = float(x[4])
    rpi = (wp*0.25) + (op*0.50) + (oop*0.25)
    data.append([int(x[0]), rpi])

variables_2015_df = pd.DataFrame(data, columns =["team_id", "team_rpi"] )
variables_2015_df.sort_values(by= "team_rpi", ascending=False)
##rpi sorted
```

```python
season_detailed_results_2015_df = pd.read_csv("regular_season_detailed_results_2015.csv")
season_detailed_results_2015_df.head()

np_detailed_results = season_detailed_results_2015_df.to_numpy()

data = []
for x in np_teams:
    steals = 0
    points = 0
    block = 0
    field_goal_made = 0
    field_goal_attempted = 0
    for y in np_detailed_results:
        if int(y[2]) == int(x):
            ## if winning team
            field_goal_made = field_goal_made + int(y[8])
            field_goal_attempted = field_goal_attempted + int(y[9])
            ## steal 18
            ## block 19
            steals = steals + int(y[18])
            block = block + int(y[19])
            points = points + int(y[3])
        if int(y[4]) == int(x):
            ## if losing team
            field_goal_made = field_goal_made + int(y[21])
            field_goal_attempted = field_goal_attempted + int(y[22])
            ## steal 18
            ## block 19
            steals = steals + int(y[31])
            block = block + int(y[32])
            points = points + int(y[5])
    if field_goal_attempted != 0 :
        data.append([x, steals, block, points, float(field_goal_made) / field_goal_attempted])
df_variables_2015 = pd.DataFrame(data, columns= ["team_id", "total_steals", "total_blocks", "total_points", "field_block_percentage"])

df_variables_2015
```
